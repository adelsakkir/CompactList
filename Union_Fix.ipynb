```python
# Candidate No: 33014

MIN = -1000
MAX = 1000

class CompactList:
    
    def __init__(self,inlist=[], c_list= None):
        self.inlist = self.merge_sort(inlist)
#         print (self.inlist)
#         print(self.find_compactlist(self.inlist))
        if c_list == None:
            self.c_list = self.find_compactlist(self.inlist)
        else:
            self.c_list = c_list

    def cardinality(self):
        cardin =0
        for i in range(1,len(self.c_list),2):
            cardin += self.c_list[i] - self.c_list[i-1]
        if len(self.c_list) % 2 != 0:
            cardin += MAX + 1 - self.c_list[-1]
        return cardin

    def insert(self,value):
        pos = 0
        for i in self.c_list:
            if value<=i:
                break
            pos+=1
        print("pos = ",pos)
        

        if pos%2==0 and pos < len(self.c_list) and self.c_list[pos]==value:
            pass #first integer of consecutive integers
        elif pos%2==0 and pos== len(self.c_list) and value <=  MAX:
            self.c_list = self.c_list[:] + [value, value +1]
        elif pos%2==0 and pos== len(self.c_list) and value >  MAX:
            pass
        elif pos%2==0 and value < self.c_list[pos]:
            
            if pos==0 and value == self.c_list[pos]-1:
                self.c_list[pos]= value
            elif pos==0:
                self.c_list = [value,value+1] + self.c_list[pos:]

            elif value == self.c_list[pos]-1:
                self.c_list[pos] = value
            
            else:
                self.c_list = self.c_list[:pos] + [value, value+1] + self.c_list[pos:]
                
        elif pos%2!=0:
            
            if pos== len(self.c_list): # and value <=  MAX:
                pass
                
            elif value == self.c_list[pos] and pos!= len(self.c_list)-1 and self.c_list[pos+1]==self.c_list[pos]+1:
                print("this")
                self.c_list.pop(pos)
                self.c_list.pop(pos)
                
            elif value == self.c_list[pos] and pos!= len(self.c_list)-1 and self.c_list[pos+1] != self.c_list[pos]+1:
                self.c_list[pos]=value+1
                
            #LAST element    
            elif value == self.c_list[pos] and pos== len(self.c_list)-1 and value != MAX:
                self.c_list[pos]=value+1
                
            elif value == self.c_list[pos] and pos== len(self.c_list)-1 and value == MAX:
                self.c_list.pop()
                
#             elif value <=  MAX:
#                 self.c_list.pop()            
            else: 
                pass #element is already in the set
        return CompactList(c_list = self.c_list)
#         print (self.c_list)
                

        
    def delete(self,value):
        pos = 0
        for i in self.c_list:
            if value<=i:
                break
            pos+=1
        print("pos = ",pos)
        
        
        if pos %2 !=0 and pos ==len(self.c_list):
            if value < MAX:
                self.c_list = self.c_list[:pos] + [value, value+1]
            elif value == MAX:
                self.c_list = self.c_list[:pos] + [value]
        elif pos %2 !=0 and self.c_list[pos] != value:
            self.c_list = self.c_list[:pos] + [value, value+1]+ self.c_list[pos:]
        elif pos %2 !=0 and self.c_list[pos] == value:
            pass
        elif pos%2 ==0 and pos == len(self.c_list):
            pass
        elif pos%2 ==0 and pos == len(self.c_list)-1:
            if self.c_list[pos] < value:
                pass
            elif self.c_list[pos] == value and value != MAX:
                self.c_list[pos] = self.c_list[pos]+1
            elif self.c_list[pos]== value and value == MAX:
                self.c_list.pop()
        elif pos%2 ==0 and self.c_list[pos]== value:
            if self.c_list[pos+1]!=self.c_list[pos]+1:
                self.c_list[pos]=self.c_list[pos]+1
            elif self.c_list[pos+1]==self.c_list[pos]+1:
                self.c_list.pop(pos)
                self.c_list.pop(pos)
        return CompactList(c_list = self.c_list)
#         print ("New List", self.c_list)
        

    def contains(self,value):
        
        #implementing binary search for to check if element in compact list
        
        if value> MAX or value < MIN:
            print("The element you are searching for is out of bounds")
            return False
        
        # if even number of elements in the list then can call binary search directly
        elif len(self.c_list) %2 ==0:
            return self.binsearch(value, 0, len(self.c_list)-1)
        
        # if odd number of elements then we check if value is greater than last element separately
        # and do binary search on remaining elements 
        elif len(self.c_list) %2 !=0 and value>=self.c_list[-1]:
            return True
        else:
            return self.binsearch(value, 0, len(self.c_list)-2)
        

    
    def binsearch(self, value, start, end):
        if start <= end:
            mid = (start + end)//2
            if mid%2 !=0:
                mid+=1
            if value >=self.c_list[mid] and value <self.c_list[mid+1]:
                return True
            elif value >= self.c_list[mid+1]:
                return self.binsearch(value, mid+2, end)
            elif value < self.c_list[mid]:
                return self.binsearch(value, start, mid-1)
        else:
            return False
    
    def subsetOf(self,cl):
        
        j=0
        i=0
        
        if len(self.c_list) %2 !=0 and len(cl) %2 ==0:
            #if cl_list goes to MAX but cl ends before max, cl_list cannot be subset of cl
            return False
        while j < len(self.c_list)-1 and i < len(cl)-1:
            if self.c_list[j] > cl[i+1]:
                i+=2
                print("case1")
            elif self.c_list[j] >= cl[i] and self.c_list[j+1] <= cl[i+1]:
                j+=2
                if j < len(self.c_list)-1 and self.c_list[j] > cl[i+1]:
                    i+=2
                print("case2")
            else :
                print("case3")
                return False
            
        if i == len(cl)-1:
            #when cl has odd elements
            if self.c_list[j] >= cl[i]:
                return True
            else:
                return False    
        elif i >= len(cl)-1 and j <len(self.c_list)-1:
            #cl is over but we still have elements in c_list
            return False
        else:
            return True 

    
    def equals(self,cl):
        
        if len(cl) != len(self.c_list):
            return False
        else:
            for i in range(len(cl)):
                if cl[i]!=self.c_list[i]:
                    return False
            return True
    

    def isEmpty(self):
        try:
            a = self.c_list[-1]
            return False
        except:
            return True

    def complement(self):
        
        if self.isEmpty():
            return CompactList(c_list = [MIN])
        
        elif len(self.c_list)%2 == 0:
            if self.c_list[0] != MIN and self.c_list[-1] != MAX:
                return CompactList(c_list = [MIN] + self.c_list[:])
            elif self.c_list[0] == MIN and self.c_list[-1] != MAX:
                return CompactList(c_list = self.c_list[1:])
            elif self.c_list[0] != MIN and self.c_list[-1] == MAX:
                return CompactList(c_list = [MIN] + self.c_list[:])
            elif self.c_list[0] == MIN and self.c_list[-1] == MAX:
                return CompactList(c_list = self.c_list[1:])

        elif len(self.c_list)%2 != 0:
            if self.c_list[0] != MIN and self.c_list[-1] != MAX:
                return CompactList(c_list = [MIN] + self.c_list[:])
            elif self.c_list[0] == MIN and self.c_list[-1] != MAX:
                return CompactList(c_list = self.c_list[1:])
            elif self.c_list[0] != MIN and self.c_list[-1] == MAX:
                return CompactList(c_list = [MIN] + self.c_list[:])
            elif self.c_list[0] == MIN and self.c_list[-1] == MAX:
                return CompactList(c_list = self.c_list[1:])
        
    def even_union(self, A, B):
        
        result = []
        l,r = 0,0

        while l < len(A) and r < len(B):
            
            # both even or both odd
            if ((l%2) + (r%2))%2 ==0 and A[l] == B[r]:
                result.append(A[l])
                l+=1
                r+=1
            #one even or one odd
            elif ((l%2) + (r%2))%2 !=0 and A[l] == B[r]:
                l+=1
                r+=1
            
            #both even
            elif l%2 ==0 and r%2 ==0:
                if A[l] < B[r]:
                    result.append(A[l])
                    l+=1
                elif A[l] > B[r]:
                    result.append(B[r])
                    r+=1
                    
            #both odd
            elif l%2 !=0 and r%2 !=0:
                if A[l] < B[r]:
                    l+=1
                elif A[l] > B[r]:
                    r+=1
            elif l%2 !=0 and r%2 ==0:
                if A[l] < B[r]:
                    result.append(A[l])
                    l+=1
                elif A[l] > B[r]:
                    r+=1
            elif l%2 ==0 and r%2 !=0:
                if A[l] < B[r]:
                    l+=1
                elif A[l] > B[r]:
                    result.append(B[r])
                    r+=1

        result.extend(A[l:])
        result.extend(B[r:])

        
        return result
    
    def union(self,cl):
        
        A = self.c_list[:]
        B = cl.c_list[:]
        
        if len(A) == 0:
            return CompactList(c_list = B)
        elif len(B) == 0:
            return CompactList(c_list = A)
            
        
        # Case1 - A and B are even
        if len(A) %2 ==0 and len(B) %2==0:
            result = self.even_union(A, B)
            return CompactList(c_list = result)
        
        #Case 2 - A is even, B is odd
        elif len(A) %2 == 0 and len(B) %2 == 1:
            while A[-2] >= B[-1]:
                A.pop()
                A.pop()
                if len(A)==0:
                    return CompactList(c_list = B[-1])
            if B[-1]>A[-1]:
                result = self.even_union(A,B[:-1])
                result.append(B[-1])
            else:
                result = self.even_union(A,B[:-1])
                result.pop()
                
            return CompactList(c_list = result)
        
        #Case 3 - A is odd, B is even
        elif len(A) %2 == 1 and len(B) %2 == 0:
            while B[-2] >= A[-1]:
                B.pop()
                B.pop()
                if len(B)==0:
                    return CompactList(c_list = A[-1])
            if A[-1]>B[-1]:
                result = self.even_union(B,A[:-1])
                result.append(A[-1])
            else:
                result = self.even_union(B,A[:-1])
                result.pop()
                

            return CompactList(c_list = result)
        
        #Case 4 - A and B are odd
        elif len(A) %2 == 1 and len(B) %2 == 1:
            if B[-1] == A[-1]:
                result = self.even_union(A[:-1], B[:-1])
                result.append(A[-1])
                
            elif B[-1] > A[-1]:
                B.pop()
                #Case 3 repeats, as B is now even
                while B[-2] >= A[-1]:
                    B.pop()
                    B.pop()
                    if len(B)==0:
                        return CompactList(c_list = A[-1])
                if A[-1]>B[-1]:
                    result = self.even_union(B,A[:-1])
                    result.append(A[-1])
                else:
                    result = self.even_union(B,A[:-1])
                    result.pop()

                return CompactList(c_list = result)
            else:
                A.pop()
                #Case 2 repeats, as A is now even
                while A[-2] >= B[-1]:
                    A.pop()
                    A.pop()
                    if len(A)==0:
                        return CompactList(c_list = B[-1])
                if B[-1]>A[-1]:
                    result = self.even_union(A,B[:-1])
                    result.append(B[-1])
                else:
                    result = self.even_union(A,B[:-1])
                    result.pop()
                
                return CompactList(c_list = result)

        

    def intersection(self,cl):
        
        A = CompactList(c_list = self.c_list).complement()
        B = cl.complement()
        AiB = (A.union(B)).complement()

        return AiB
        

    def difference(self,cl):
        
        #A is an object, B is a compactlist
        A = CompactList(c_list = self.c_list)
        B = cl.complement()
        print("A - ", A.c_list)
        print("B - ", B.c_list)
        return A.intersection(B)
            
    def __str__(self):
        
        if len(self.c_list) ==0:
            return("[]")
        
        output ="[" + str(self.c_list[0]) +","
        
        for i in range(1,len(self.c_list)):
            if i%2 ==0:
                output += "] U ["
                output += str(self.c_list[i])
                output+= ","
            else:
                output += str(self.c_list[i]-1)
                
        if len(self.c_list) %2 ==0:
            output +="]"
        else:
            output += str(MAX) +"]" 
        return(output)
    
    def merge(self, left, right):
        l,r=0,0
        result = []
        while l<len(left) and r<len(right):
            if left[l]<=right[r]:
                result.append(left[l])
                l+=1
            else:
                result.append(right[r])
                r+=1
        result+=left[l:]
        result+=right[r:]

        return result

    def merge_sort(self, array):
        
        if len(array)==0:
            return []
        elif len(array)==1:
            return array

        mid=len(array)//2
        left = self.merge_sort(array[:mid])
        right= self.merge_sort(array[mid:])

        return self.merge(left,right)
    
    def find_compactlist(self, array):
        
        len_array = len(array)
        if len_array == 0:
            return []
        elif array[0] < MIN or array[-1] > MAX:
            print("Elements of the list entered are beyond the range MAX-MIN, not compatible")
            return []
        else:
            c_list = [array[0]]
            for i in range(1,len_array):
                if array[i] > array[i-1]+1:
                    c_list.append(array[i-1]+1)
                    c_list.append(array[i])
        if array[-1] != MAX:
            c_list.append(array[-1]+1)
            
        return c_list

if __name__ == "__main__":

    mycl1 = CompactList([])
    print(mycl1)
    print(mycl1.cardinality())
    
    mycl2 = CompactList([9,8,3,4,5])
    print(mycl2)
    print(mycl2.cardinality())
    print(mycl2, "contains 4:",mycl2.contains(4)) 
    print("Delete - ")
    print(mycl2.delete(9))
    print("Deleted C_list",mycl2)
    print("Deleted C_list",mycl2.c_list)
    
#     print("Insert - ")
#     print ("Before Insertion 7 -", mycl2)
#     print(mycl2.insert(7))
#     print ("After Insertion -", mycl2)
#     print("Insert C_list",mycl2.c_list)

    mycl1 = CompactList([10,1,11])
    mycl3 = mycl1.union(mycl2)
    print(mycl3)
    print(mycl3.cardinality())
    
    mycl1 = mycl2.complement()
    print(mycl1)
    print(mycl1.cardinality())

    mycl1 = CompactList([MAX])
    print(mycl1)
    print(mycl1.cardinality())
```
